OUTLIERS

```python

Q1 = df['MedInc'].quantile(0.25)
Q3 = df['MedInc'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

outliers = df[(df['MedInc'] < lower_limit) | (df['MedInc'] > upper_limit)]

print("Number of outliers in MedInc:", outliers.shape[0])
print("Lower limit:", lower_limit)
print("Upper limit:", upper_limit)
```

```python

Q1 = df['HouseAge'].quantile(0.25)
Q3 = df['HouseAge'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

outliers = df[(df['HouseAge'] < lower_limit) | (df['HouseAge'] > upper_limit)]

print("Number of outliers in HouseAge:", outliers.shape[0])
print("Lower limit:", lower_limit)
print("Upper limit:", upper_limit)
```

```python
outlier_summary = {}
lower_limits_dict = {}
upper_limits_dict = {}

for col in df.columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR


    outliers = df[(df[col] < lower_limit) | (df[col] > upper_limit)]
    outlier_summary[col] = len(outliers)
    lower_limits_dict[col] = lower_limit
    upper_limits_dict[col] = upper_limit


outlier_df = pd.DataFrame.from_dict(outlier_summary, orient='index', columns=['Outlier_Count'])
outlier_df['Total_Rows'] = len(df)
outlier_df['Lower_Limit'] = pd.Series(lower_limits_dict)
outlier_df['Upper_Limit'] = pd.Series(upper_limits_dict)
outlier_df['Outlier_%'] = (outlier_df['Outlier_Count'] / outlier_df['Total_Rows']) * 100

print("\nOutlier count and percentage per column:")
print(outlier_df)
```

```python
x=df
y=data.target
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
print("x_train:",len(x_train))
print("x_test:",len(x_test))
print("y_train:",len(y_train))
print("y_test:",len(y_test))
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)

print(f"Linear regession Accuracy: {lr.score(x_test, y_test)}")
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_train,y_train)

print(f"Random forest Accuracy: {rf.score(x_test, y_test)}")
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
print(f"Decision tree Accuracy: {dt.score(x_test, y_test)}")

```

```python
from sklearn.linear_model import  Lasso



lasso_model = Lasso()
lasso_model.fit(x_train, y_train)

print(f"lasso R2 score: {lasso_model.score(x_test, y_test)}")
```

```python
from sklearn.linear_model import  Ridge



ridge_model = Ridge()
ridge_model.fit(x_train, y_train)
print(f"Ridge R2 score: {ridge_model.score(x_test, y_test)}")
```

```python
from sklearn.metrics import r2_score

models = [
    ('Linear Regression', lr),
    ('Random Forest', rf),
    ('Decision Tree', dt),
    ('Lasso', lasso_model),
    ('Ridge', ridge_model)
]

model_names = []
r2_scores = []

for name, model in models:
    y_pred = model.predict(x_test)
    score = r2_score(y_test, y_pred)
    model_names.append(name)
    r2_scores.append(score)
    print(f"{name} R2 Score: {score:.4f}")

import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
bars = plt.bar(model_names, r2_scores, color=['blue', 'green', 'orange', 'red', 'purple'])
plt.ylim(0, 1)
plt.title("Regression Model R2 Score Comparison")
plt.ylabel("R2 Score")
plt.xlabel("Model")

plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error


y_pred_lr = lr.predict(x_test)
y_pred_rf = rf.predict(x_test)
y_pred_dt = dt.predict(x_test)
y_pred_lasso = lasso_model.predict(x_test)

y_pred_ridge = ridge_model.predict(x_test)


models = ['Linear Regression', 'Random Forest', 'Decision Tree','Lasso', 'Ridge']

mse_scores = [
    mean_squared_error(y_test, y_pred_lr),
    mean_squared_error(y_test, y_pred_rf),
    mean_squared_error(y_test, y_pred_dt),
    mean_squared_error(y_test, y_pred_lasso),

    mean_squared_error(y_test, y_pred_ridge)
]

mae_scores = [
    mean_absolute_error(y_test, y_pred_lr),
    mean_absolute_error(y_test, y_pred_rf),
    mean_absolute_error(y_test, y_pred_dt),
    mean_absolute_error(y_test, y_pred_lasso),
    mean_absolute_error(y_test, y_pred_ridge)
]

# Display scores
for i, model in enumerate(models):
    print(f"{model} -> MSE: {mse_scores[i]:.4f}, MAE: {mae_scores[i]:.4f}")
```
